### Import Libraries and Dependencies

```python
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
```

### Read data

```python
portfolio_a_path = Path("../Resources/tech_stocks_2018_a.csv")
portfolio_b_path = Path("../Resources/tech_stocks_2018_b.csv")

portfolio_a = pd.read_csv(portfolio_a_path, index_col='Date', parse_dates=True, infer_datetime_format=True)
portfolio_b = pd.read_csv(portfolio_b_path, index_col='Date', parse_dates=True, infer_datetime_format=True)
portfolio_a.head()
```

```python
portfolio_a.sort_index(ascending=True, inplace=True)
portfolio_b.sort_index(ascending=True, inplace=True)
```

### Calculate Annualized Std Dev

```python
# Calculate daily returns
daily_return_a = portfolio_a.pct_change()
daily_return_b = portfolio_b.pct_change()

# Concat returns into one DataFrame
daily_return = pd.concat([daily_return_a, daily_return_b], axis="columns", join="inner")
daily_return
```

```python
# Calculate Sharpe Ratio
daily_std = daily_return.std()
annualized_return = daily_return.mean() * 252
annualized_std = daily_return.std() * np.sqrt(252)
sharpe_ratio = annualized_return / annualized_std
sharpe_ratio
```

### Plot sharpe ratios

```python
# Plot sharpe ratios
sharpe_ratio.plot(kind="bar", title="Sharpe Ratios")
```

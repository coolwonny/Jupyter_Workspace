# World Happiness

## 1. Import Libraries and Dependencies

```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

## 2. Set File Path to CSV using Pathlib

```python
# Create reference to CSV file
# Hint - the file path may require some extra steps..
csvpath = Path('../Resources/world_happiness.csv')
```

## 3. Import CSV into Pandas DataFrame

```python
# Import the CSV into a pandas DataFrame
df = pd.read_csv(csvpath)
df.head()
```

## 4. Set Index and Sort Index in Ascending Order

```python
# Set new index to Country
df = df.set_index('Country')
# Sort the new dataframe alphabetically
df = df.sort_index()
```

## 5. Select the Happiness Score for Vietnam

```python
# Grab the data contained within the row of a country of your choice and the "Happiness.Score" column
# Print the result as "<country's name> happiness score for 2017 was <happiness score>"
vietnam = df.loc['Vietnam', 'Happiness.Score']
print(f'Vietnam happiness score for 2017 was {round(vietnam, 3)}')
```

## 6. Select the Happiness Rank and Score for Belgium, Belize, Benin, Bhutan, and Bolivia

```python
# Grab all rows of data from "Belgium" to "Bolivia" and the columns from "Happiness.Rank" to "Happiness.Score"
# Hint - in order to view more data you will need to increase the dataframe's head
df.head(20)
df.loc['Belgium':'Bolivia', 'Happiness.Rank':'Happiness.Score']
```

## 7. Select the Freedom and Government Trust for Every Country

```python
# Select all rows and only display the columns for "Freedom" and "Trust..Government.Corruption." 
df.loc[:, ['Freedom', 'Trust..Government.Corruption.' ]]
```

## 8. Select all the Countries Where GDP is Greater than or Equal to 1.6

```python
# Select all rows where the GDP per capita has a value greater than or equal to 1.6
df.loc[df['Economy..GDP.per.Capita.'] >= 1.6]
```

## 9. Select all the Countries Where the Happiness Score is Greater than or Equal to 5 and GDP is Less than or Equal to 1

```python
# Select all rows that have a happiness score greater than or equal 5 and a GDP score less than or equal 1
df.loc[df['Economy..GDP.per.Capita.'] <= 1.0].loc[df['Happiness.Score'] >= 5]
```
